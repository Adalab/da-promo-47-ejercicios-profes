## Preparación de los datos: 

```python
# poned todos los nombres de las columnas en minúsculas

nuevas_cols = {col: col.lower() for col in df.columns}
df.rename(columns=nuevas_cols, inplace=True)

df.columns = df.columns.str.lower()
```

```python
# haciendo el análisis exploratorio nos deberíamos haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que debería. Esto es debido a que los decimales están establecidos como comas
# y no con puntos. Cread una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto

def cambiar_comas(x):
    try:
        return float(x.replace(",", "."))
    except:
        return np.nan
```

```python
lista_columnas = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']

for i in lista_columnas:
    df[i] = df[i].apply(cambiar_comas)
```

```python
# después de haber hecho los cambios, chequead los tipos de datos. ¿Son ya del tipo correcto? En caso de que no, cread otra función o modificar la anterior para que sean de tipo float. YA ESTA SOLUCIONADO EN LA FUNCIÓN. 
df.dtypes
```

```python
# hay algunas canciones cuyo título está en mayúsulas. Crea una función para que todos los títulos estén en minúscula. Haced lo mismo para las columnas de "track" y "album"

df["title"] = df["title"].apply(lambda x: x.lower() if type (x) == str else x)
df["track"] = df["track"].apply(lambda x: x.lower() if type (x) == str else x)
df["album"] = df["album"].apply(lambda x: x.lower() if type (x) == str else x)

df.tail()
```

```python
# algunos de los nombres de los artistas tienen símbolos raros que nos pueden dar errores a la hora de hacer visualizaciones. Lo primero que deberéis hacer es explorar los nombres de los artistas e identificar estos caracteres raros. Una vez identificados, quitad esos caracteres de esa columna. 

df["artist"].unique().tolist()[:10]
```

```python
# usando el dataframe de los albumes que hemos creado en el ejercicio anterior. ¿Cuál es la media y la desviación estándar de "danceability", "acousticness" y "speechiness" de cada artista?. Mostrad los resultados en tres dataframes diferentes, uno para cada métrica
# ordenar los resultados de mayor  a menor en base a la media. 
metricas = ["danceability", "acousticness" , "speechiness"]
for metrica in metricas: 
    df_artistas = df.groupby("artist")[metrica].agg(["mean", "std"]).reset_index()
    df_artistas.sort_values(by = "mean", ascending=False)
    print(metrica)
    display(df_artistas.head())
```

```python
# ¿Son aquellos artistas que tienen más visitas los que más likes tienen? 
# Para solucionar este ejercicio deberéis, agupar por artista y calcular la media de visitas y de likes. 

df_likes = df.groupby("artist")["views", "likes"].mean().reset_index()
df_likes.head()
```

```python
# quedaros con los 10 artistas que más visitas han tenido

df_top_views = df_likes.sort_values(by="views", ascending=False).head(10)
df_top_views
```

```python
# quedaros con los 10 artistas que más likes han tenido

df_top_likes = df_likes.sort_values(by="likes", ascending=False).head(10)
df_top_likes
```
