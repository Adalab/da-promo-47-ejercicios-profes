## loc/iloc para un valor
- .loc para etiquetas
- .iloc para posiciones numericas, para indexar por posiciones
- [filas-deseadas, :]  le estamos diciendo que nos busque por una fila dada y nos devuelva todas las columnas
-  [:, columna-deseada]  devuelvenos todas las filas de la columna que especifiquemos
-  df.loc[df["columna"] CONDICION]

```python
## indices
print(df.iloc[0,1])

## etiquetas
print(df.loc["Mon", "Temperature"])
```

```python
# informacion climatica del martes
# indices
print(df.iloc[1,:])
#etiquetas
print(df.loc['Tue',:])
```

```python
## viento de todos los dias
print(df.iloc[:,2])
print(df.loc[:,'Wind'])
```

```python
# humedad del martes y miercoles 
print(df.loc[['Tue', 'Wed'], 'Humidity'])

# humedad del martes y miercoles con index
print(df.iloc[[1,2], [2,3]])
```

```python
# filas ue la temperatura es mayor que 12
df.loc[df['Temperature']> 12, :]
```

```python
## dias soleado y poco viento 
df.loc[ (df['Weather']=='Sunny') & (df['Wind']> 12), :]
```

## Series

- son unidimensionales
- mismo tipo de datos
- tienen un índice
- características de las series
  - .values --> nos devuelve los valores de la serie
  - .size -> numero total de elementos

## Usar listas vs Series
- Lista: Es una colección de elementos ordenados que no tienen un índice explícito asociado (más allá de su posición).
- Serie: Es una estructura de datos con un índice asociado a cada elemento, lo que permite acceder a los datos tanto por posición como por etiqueta.
- Usa listas cuando solo necesites una lista básica de cosas.
- Usa Series cuando necesites organizar datos con nombres o hacer cálculos rápidos

```python
## crear series: 
serie_vacia = pd.Series()
serie_vacia

lista = [23, 45, 67, 89, 76]
serie = pd.Series(lista)
serie

serie = pd.Series(range(2,9))
serie
```

```python
serie.index = range(9,16)
serie
```

```python
# Crear la serie con valores de 2 a 8 (7 elementos)
serie = pd.Series(range(2, 9))

# Cambiar los índices a partir de 9 hasta el final dinámicamente
serie.index = range(9, 9 + len(serie))

print(serie)

### range(9, 9 + len(serie)) genera un rango que empieza en 9 y tiene el mismo tamaño que la serie.
# len(serie) se usa para asegurarnos de que el rango tenga la cantidad exacta de índices.
```

```python
dias = ['lunes', 'martes', 'miercoles', 'jueves', 'viernes', 'sabado', 'dom']
serie.index = dias
serie
```

```python
print(serie.values) 
print(serie.size)
print(serie.index)
print(serie.shape)
```

```python
serie["jueves"]
serie[5]
serie[[0, 2, 3]]


# iloc /loc
serie.loc['lunes']
serie.loc[['lunes', 'martes']]

serie.iloc[4]
serie.iloc[1:4]
```
