# Gestión de nulos

```python
## ver los nulos
#static_fire_date_utc
## ver la cantidad de nulos
df.isnull().sum()[df.isnull().sum()>0]/df.shape[0]
```

```python
## columnas qye tengan nulos y sean catgoricas
nulos_categ =  df[df.columns[df.isna().any()]].select_dtypes(include = ["object", 'datetime64']).columns
nulos_categ
```

```python
## columnas que tengan nulos pero que sean numericas
nulos_num =  df[df.columns[df.isnull().any()]].select_dtypes(include = np.number).columns
nulos_num =  df[df.columns[df.isnull().any()]].select_dtypes(include = [int, float]).columns
nulos_num
```

```python
## imputar un variable numerica
## cantidad de nulos: 42% 

imputer_iter = IterativeImputer(max_iter = 10, random_state = 42)
df['window'] = imputer_iter.fit_transform(df[['window']])

```
